```python
from bs4 import BeautifulSoup
```

```python
url='http://www.cricmetric.com/ipl/ranks/'
```

```python
import requests
```

```python
response=requests.get(url)
```

```python
response
```

```python
soup=BeautifulSoup(response.text)
```

```python
soup
```

```python
team=soup.select(".scoretable td~ td+ td a")
```

```python
players=soup.select(".scoretable td:nth-child(2) a")
```

```python
rank=soup.select(".scoretable td:nth-child(1)")
```

```python
import pandas
```

```python
ipl_team=pandas.DataFrame(rank,columns=['Rank'])
```

```python
ipl_team
```

```python
ipl_team["Player Name"]=players
```

```python
ipl_team["team name"]=team
```
